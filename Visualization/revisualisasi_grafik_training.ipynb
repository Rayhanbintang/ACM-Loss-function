```python
loss_dict = {'wxent': 'Weighted Cross Entropy',
           'xent': 'Cross Entropy',
           'onlyfg': 'Dice Only-Foreground',
           'dice': 'Dice',
            'ac':'Active Contour'
           }
graph_dict = {'training_dice': "Training Dice",
             'training_loss':'Training Loss',
             'validation_dice':'Validation Dice',
             'validation_loss':'Validation Loss',
             'loss': 'Loss'}

def get_loss_name(filename):    
    loss_name = filename.split('-')[0]
    loss_name = loss_name.split('_')[-1]
    
    return loss_dict[loss_name]

def get_graph_name(filename): 
    graph_name = filename.split('-')[2]
    graph_name = graph_name.split('.')[0]
    return graph_dict[graph_name]
```
