```python
import os
dgx = 'local_bs_4'
path = os.path.join(os.getcwd(), dgx)

files = os.listdir(path)
```

```python
loss_dict = {'wxent': 'Weighted Cross Entropy',
           'xent': 'Cross Entropy',
           'onlyfg': 'Dice Only-Foreground',
           'dice': 'Dice',
            'ac':'Active Contour'
           }
graph_dict = {'training_dice': "Training Dice",
             'training_loss':'Training Loss',
             'validation_dice':'Validation Dice',
             'validation_loss':'Validation Loss',
             'loss': 'Loss'}

def get_loss_name(filename):    
    loss_name = filename.split('-')[0]
    loss_name = loss_name.split('_')[-1]
    
    return loss_dict[loss_name]

def get_graph_name(filename): 
    graph_name = filename.split('-')[2]
    graph_name = graph_name.split('.')[0]
    return graph_dict[graph_name]
```

```python
# enumerate file di path biar gampang ganti-ganti file
for i, v in enumerate(files):
    print(i, v)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

#ganti angka ini berdasarkan angka dari cell diatas 
# 8 loss
# 9 training_dice
# 10 training_loss
# 11 validation_dice
# 12 validation_loss

num1 = 9
num2 = num1 + 5
num3 = num2 + 5
num4 = num3 + 5
num5 = num4 + 5


df1 = pd.read_csv(os.path.join(path, files[num1]))
df1_loss = get_loss_name(files[num1])
df1_gr = get_graph_name(files[num1])

print ('grafik', df1_gr)
print(df1_loss)

df2 = pd.read_csv(os.path.join(path, files[num2]))
df2_loss = get_loss_name(files[num2])
df2_gr = get_graph_name(files[num2])
print(df2_loss)
print (df2_gr)

df3 = pd.read_csv(os.path.join(path, files[num3]))
df3_loss = get_loss_name(files[num3])
df3_gr = get_graph_name(files[num3])
print(df3_loss)
print (df3_gr)

df4 = pd.read_csv(os.path.join(path, files[num4]))
df4_loss = get_loss_name(files[num4])
df4_gr = get_graph_name(files[num4])
print(df4_loss)
print (df4_gr)
```

```python
fontsize = 26
plt.figure(figsize=(16,9))
plt.title(df1_gr, fontsize=fontsize)
plt.plot(df1.Step, df1.Value, label=df1_loss)
plt.plot(df2.Step, df2.Value, label=df2_loss)
plt.plot(df3.Step, df3.Value, label=df3_loss)
plt.plot(df4.Step, df4.Value, label=df4_loss)

plt.ylim(top=1, bottom=-0.2)
plt.xlabel("Step", fontsize=fontsize)
plt.ylabel("Dice", fontsize=fontsize)
plt.legend(loc='lower center',prop={'size': fontsize} )
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.grid(True)
plt.show
```
