```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_excel('2D.xlsx')
df.columns
```

```python
df_dice = df.loc[:, ['AC_L1', 'AC_L5', 'AC_L10', 'WCE', 'CE', 'DICE', 'DOF']]

fontsize=26
plt.figure(figsize=(16,9))
sns.boxplot(x="variable", y="value", data=pd.melt(df_dice))
plt.xlabel("Loss Function", fontsize=fontsize)
plt.xticks(fontsize=15)
# plt.legend(fontsize=fontsize)
plt.ylabel("Dice Score", fontsize=fontsize)
plt.title("Pengukuran Dice secara 2D", fontsize=fontsize)
```
